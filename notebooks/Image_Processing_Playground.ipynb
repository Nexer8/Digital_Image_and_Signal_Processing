### Libraries used

```python
import numpy as np
from scipy import misc, ndimage
from skimage.exposure import equalize_hist
from skimage import data, filters
from matplotlib.pyplot import subplot, title, show, imshow, hist, rcParams, subplots_adjust
%matplotlib inline
rcParams['figure.figsize'] = [10, 8]
```

### Automatic equalization of the histogram consisting in improving the contrast of the image

```python
# Image equalization
image = data.coins()
equalized = np.asarray(equalize_hist(image) * 255, dtype='uint8')

subplot(221), imshow(image, cmap='gray'), title('Original')
subplot(222), hist(image.flatten(), 256, range=(
    0, 256)), title('Histogram of original')
subplot(223), imshow(equalized, cmap='gray'), title('Enhanced contrast')
subplot(224), hist(equalized.flatten(), 256, range=(
    0, 256)), title('Histogram of equalilzed')

subplots_adjust(hspace=1.02)
show()
```

### Using the Sobel filter used for edge detection

```python
# Sobel filter usage
image = data.coins()
filtered = filters.sobel(image)

subplot(121), imshow(image, cmap='gray'), title('Original')
subplot(122), imshow(filtered, cmap='gray'), title(
    'Filtered with Sobel filter')
show()
```

### Using the Prewitt filter used for edge detection

```python
# Prewitt filter usage
image = data.coins()
filtered = filters.prewitt(image)

subplot(121), imshow(image, cmap='gray'), title('Original')
subplot(122), imshow(filtered, cmap='gray'), title(
    'Filtered with Prewitt filter')
show()
```

### Applying a mask to an image with a predetermined brightness

```python
# Masking an image
image = data.camera()

subplot(121), imshow(image, cmap='gray'), title('Original')
mask = image < 100
image[mask] = 255

subplot(122), imshow(image, cmap='gray'), title('Masked image')
show()
```

### Using Gauss averaging filters with different degree of blurring

```python
# Gaussian filter
image = misc.face(gray=True)
gaussian3 = ndimage.gaussian_filter(image, sigma=3)
gaussian5 = ndimage.gaussian_filter(image, sigma=5)

subplot(131), imshow(image, cmap='gray'), title('Original')
subplot(132), imshow(gaussian3, cmap='gray'), title('Gaussian, sigma=3')
subplot(133), imshow(gaussian5, cmap='gray'), title('Gaussian, sigma=5')
show()
```
