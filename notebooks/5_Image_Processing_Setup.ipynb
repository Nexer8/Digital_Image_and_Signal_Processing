### Task 1

Write a script in *Python* / *Matlab* to load and visualize the examined images. The program should enable displaying an image loaded from a file with a given name, selecting a sub-image (rectangular area) with given coordinates and saving it to a given file.

### Libraries used

```python
import os
import skimage as sk
from skimage import io, img_as_uint
from matplotlib.pyplot import imshow, title, show, savefig
```

### The actual script that saves both the figure generated by matplotlib.pyplot and the image itself

```python
image = io.imread(os.path.join(sk.data_dir, 'coins.png'))

imshow(image, cmap='gray'), title('Original image')
io.imsave('../data/images/' + 'coins.png',
          img_as_uint(image))    # save the image
savefig('../data/images/' + 'coins_fig.png')     # save the figure
show()
```
