Try the histogram equalization action on the sample images. To see the effectiveness of the procedure, smooth out images that are too dark and too light. Draw histograms of the images before and after equalization.

#### Libraries used

```python
import numpy as np
from skimage.exposure import equalize_hist
from matplotlib.pyplot import subplot, imshow, title, show, hist, rcParams
from skimage import io
import skimage as sk
import os
%matplotlib inline
rcParams['figure.figsize'] = [20, 10]
```

#### Loading a low contrast image

```python
low_contrast_image = io.imread(os.path.join(
    sk.data_dir, 'moon.png'))   # low contrast image
```

#### Loading a high contrast image

```python
high_contrast_image = io.imread(
    '../data/images/too_bright.png')   # high contrast image
```

#### Equalizing the contrast for both images

```python
low_contrast_equalized = np.asarray(
    equalize_hist(low_contrast_image) * 255, dtype='uint8')
high_contrast_equalized = np.asarray(
    equalize_hist(high_contrast_image) * 255, dtype='uint8')
```

#### Displaying the results

```python
subplot(241), imshow(low_contrast_image,
                     cmap='gray'), title('Low contrast image')
subplot(242), hist(low_contrast_image.flatten(), 256, range=(0, 256))

subplot(243), imshow(high_contrast_image,
                     cmap='gray'), title('High contrast image')
subplot(244), hist(high_contrast_image.flatten(), 256, range=(0, 256))

subplot(245), imshow(low_contrast_equalized,
                     cmap='gray'), title('Equalized contrast')
subplot(246), hist(low_contrast_equalized.flatten(), 256, range=(0, 256))

subplot(247), imshow(high_contrast_equalized,
                     cmap='gray'), title('Equalized contrast')
subplot(248), hist(high_contrast_equalized.flatten(), 256, range=(0, 256))
show()
```
