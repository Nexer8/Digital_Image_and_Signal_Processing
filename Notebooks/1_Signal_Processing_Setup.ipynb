Write a script in Python / Matlab that allows you to load and visualize the analyzed signals.

The program should enable observation of a signal slice for a given time interval, scaling the graph axes and their description, and saving any signal slice to a file with a given name.

```python
import matplotlib.pyplot as plt
import numpy as np

frequency = 1000  # set the frequency to 1000 Hz

data = np.loadtxt('../data/ekg1.txt', delimiter=' ')
ecg_data = np.hsplit(data, 12)

time = np.arange(len(ecg_data[0])) / frequency  # Create a set of time measurements

for i, ecg in enumerate(ecg_data):
    plt.rcParams['figure.figsize']=[10, 8]
    plt.plot(time, ecg)
    plt.xlabel('time [s]')
    plt.ylabel('ECG [mV]')
    # plt.xlim(0, 0.1)  # select just a part of x axis to render
    # plt.ylim(0, 200)  # select just a part of y axis to render
    plt.savefig(f'../results/graph{i}.png')
    plt.show()
```
